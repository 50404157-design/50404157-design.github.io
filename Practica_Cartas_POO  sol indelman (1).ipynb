Conocimiento de base: Un mazo de cartas españolas trae 50 cartas. Estas estan clasificadas segun su _palo_, que puede ser Bastos, Espadas, Copas u Oros. Hay 12 cartas de cada tipo, numeradas correspondientemente. Es común llamar Sota a la carta con el número 10, Caballo a la carta con el número 11, Rey a la carta con el número 12 y As a la carta con el número. El mazo de cartas de españolas se completa con dos comodines.

**Ejercicio 1**: Definir una clase Carta que contenga todos los atributos necesarios para describir las 40 cartas españolas necesarias para el Truco.


```python
# Escribí tu código acá

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo

    def mostrar_carta(self):
        nombres = {
            1: "As",
            10: "Sota",
            11: "Caballo",
            12: "Rey"
        }

        nombre = nombres.get(self.numero, str(self.numero))
        print(f"{nombre} de {self.palo}")


# Casos de prueba
carta1 = Carta(1, "Espadas")
carta2 = Carta(7, "Oros")
carta3 = Carta(10, "Copas")
carta4 = Carta(12, "Bastos")

carta1.mostrar_carta()
carta2.mostrar_carta()
carta3.mostrar_carta()
carta4.mostrar_carta()
```

**Ejercicio 3**: Extender la definicion de Carta para definir un método que nos permita imprimir las cartas como lo haríamos naturalmente. Ejemplo: 5 de Basto.

```python
class Carta:

    apodos = {
        1: "Ancho",
        10: "Sota",
        11: "Caballo",
        12: "Rey"
    }

    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo

    # Método para imprimir la carta naturalmente
    def __str__(self):
        nombre = Carta.apodos.get(self.numero, str(self.numero))
        return f"{nombre} de {self.palo}"


# Casos de prueba
ancho_de_espadas = Carta(1, "Espada")
cinco_de_basto = Carta(5, "Basto")
rey_de_copas = Carta(12, "Copas")

print(ancho_de_espadas)
print(cinco_de_basto)
print(rey_de_copas)
```

**Ejercicio 4**: Escribir un método que nos permita comparar cartas por igualdad

```python
class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo

    def __str__(self):
        return f"{self.numero} de {self.palo}"

    # Método para comparar igualdad
    def __eq__(self, other):
        return self.numero == other.numero and self.palo == other.palo


# Casos de prueba
carta1 = Carta(1, "Espadas")
carta2 = Carta(1, "Espadas")
carta3 = Carta(7, "Oros")

print(carta1 == carta2)  # True
print(carta1 == carta3)  # False
```

**Ejercicio 5**: Escribir una clase mazo, que construya el mazo de cartas españolas. Escribir un método que devuelve cuantas cartas hay en el mazo.

```python
class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo

    def __str__(self):
        return f"{self.numero} de {self.palo}"


class Mazo:
    def __init__(self):
        self.cartas = []

        palos = ["Bastos", "Espadas", "Copas", "Oros"]

        # Crear las 48 cartas españolas
        for palo in palos:
            for numero in range(1, 13):
                self.cartas.append(Carta(numero, palo))

        # Agregar comodines
        self.cartas.append(Carta("Comodín", ""))
        self.cartas.append(Carta("Comodín", ""))

    def cantidad(self):
        return len(self.cartas)


# Casos de prueba
mazo = Mazo()

print("Cantidad de cartas en el mazo:", mazo.cantidad())
```

**Ejercicio 6**: Escribir un método en la clase Mazo que *mezcle* el mazo. Puede ser de utilidad el módulo [`random`](https://docs.python.org/3/library/random.html) de la biblioteca estándar de Python.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo

    def __str__(self):
        return f"{self.numero} de {self.palo}"


class Mazo:
    def __init__(self):
        self.cartas = []

        palos = ["Bastos", "Espadas", "Copas", "Oros"]

        # Crear las cartas españolas
        for palo in palos:
            for numero in range(1, 13):
                self.cartas.append(Carta(numero, palo))

        # Agregar comodines
        self.cartas.append(Carta("Comodín", ""))
        self.cartas.append(Carta("Comodín", ""))

    def cantidad(self):
        return len(self.cartas)

    # Método para mezclar el mazo
    def mezclar(self):
        random.shuffle(self.cartas)


# Casos de prueba
mazo = Mazo()

print("Antes de mezclar:")
print(mazo.cartas[0])
print(mazo.cartas[1])

mazo.mezclar()

print("\nDespués de mezclar:")
print(mazo.cartas[0])
print(mazo.cartas[1])
```
